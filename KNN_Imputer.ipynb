# KNN Imputer

```python
import numpy as np
import pandas as pd
from sklearn.impute import KNNImputer,SimpleImputer

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("train.csv",usecols=["Age","Pclass","Fare","Survived"])
df.head(2)
```

```python
df.isnull().sum()
```

```python
x = df.drop(columns=["Survived"])
y = df["Survived"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
knn = KNNImputer()
x_train_knn = knn.fit_transform(x_train)
x_test_knn = knn.transform(x_test)
```

```python
pd.DataFrame(x_train_knn,columns=x_train.columns)
```

```python
lr = LogisticRegression()
lr.fit(x_train_knn,y_train)
y_pred = lr.predict(x_test_knn)
accuracy = accuracy_score(y_test,y_pred)
print(f"KNN Imputer Acuuracy: {accuracy*100:.2f}%")
```

```python
si = SimpleImputer()
x_train_si = si.fit_transform(x_train)
x_test_si = si.transform(x_test)
```

```python
lr = LogisticRegression()
lr.fit(x_train_si,y_train)
y_pred = lr.predict(x_test_si)
accuracy = accuracy_score(y_test,y_pred)
print(f"Simple Imputer Acuuracy: {accuracy*100:.2f}%")
```
