# 2. Ordinal Encoding

<!-- 2. Ordinal encoding -->

```python
df = pd.read_csv("covid_toy.csv")
df.head()
```

```python
df.dropna()
```

```python
df = df.drop(columns=['age','fever','city'])
```

```python
df.head(2)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Male','Female'],
                               ['Mild','Strong'],
                               ['No','Yes']])
```

```python
oe.fit_transform(df)
```

# 3. OneHotEncoding

```python
df = pd.read_csv("covid_toy.csv")
df.head(2)
```

```python
df = df.dropna()
```

```python
from sklearn.preprocessing import OneHotEncoder 
```

```python
ohe = OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
```

```python
a = ohe.fit_transform(df[['gender','cough','city','has_covid']])
a
```

```python
a.shape
```

# 4. get_dummies

```python
df = pd.read_csv("covid_toy.csv")
df = df.dropna()
```

```python
df
```

```python
a = pd.get_dummies(df, columns = ['gender', 'cough','city','has_covid'], drop_first = True)
```
